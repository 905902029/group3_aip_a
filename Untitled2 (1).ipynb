```
def pic(hp):
  k=""
  if(int(hp)>6):
    k="\n|      |"
    add_times = int(hp)-6
    for x in range(add_times):
      k = k+k
  p="_________ "+k+"""         
|      |     
|      0     
|     /|\    
|     / \    
|            
"""
  STAGES = p.rstrip().splitlines()
  return STAGES



def hangman(name, word,STAGES):
    text="ハングマンへようこそ!"
    add_times = 100
    for x in range(add_times):
        text = "**************************************************************\n"+text 
    print(text)
    rletters = list(word)
    board = ["_"] * len(word)
    hung = 1
    while hung < len(STAGES):
        char = input_char(name)
        if char in rletters:
            index = rletters.index(char)
            board[index] = char
            rletters[index] = '$'
        else:
            hung += 1
        print(*board)
        if "_" not in board:
            print("あなたの勝ち!")
            return
        print(*STAGES[:hung], sep="\n")
    print(f"あなたの負け!正解は {word}.")


def input_char(name):
    while True:
        char = input(f"\n{name}: 1文字(英字小文字)を予想してね! ")
        if not char.islower():
            print("英小文字を入力してね")
        elif len(char) != 1:
            print("1文字ずつだよ!")
        else:
            return char


def input_word(name):
    while True:
        word = input(f"{name}: 英単語を入力してね!(3文字以上6文字以内)")
        if not word.islower():
            print("英小文字を入力してね")
        elif not 3 <= len(word) <= 6:
            print("3文字以上6文字以内だよ!")
        else:
            return word
def input_hp(name):
    while True:
        hp = input(f"{name}: 数字を入力してね!(6-26以内)")
        if not hp.isnumeric():
            print("数字を入力してね")
        elif not 6 <= int(hp) <= 26:
            print("6-26以内だよ!")
        else:
            return hp

if __name__ == '__main__':
    hangman("player2", input_word("player1"),pic(input_hp("player1のHP")))
```

```
#introuce random
import random
import os

#建立随机单词库
dictionary=("cat","dog","rabbit","bear","sheep")
def hangman(word): #定义函数
    wrong = 0#错误的次数
    HP = ["",
              "_______      ",
              "|            ",
              "|      |     ",
              "|      0     ",
              "|     /|\    ",
              "|     / \    ",
              "|            "
              ]
    rletters = list(word)#释放列表WORD
    board = ["_"] * len(word)#根据随机到的单词长度展现下划线，
    win = False#默认是失败
    print("欢迎来到猜单词(动物)!")
    while wrong < len(HP) - 1:#如果wrong的次数小于生命值就会继续运行
        print("\n")
        msg = "请输入字母! "
        char = input(msg)#请输入字母
        if char in rletters:#判断机制，如果字母正确
            cind = rletters.index(char)#判断单词位置
            board[cind] = char#替换字母
        else:#错误计数器 如果错误计数器就加1
            wrong += 1
        print(" ".join(board))#不管正确还是错误都打印单词板
        e = wrong + 1
        print("\n".join(HP[0:e]))#记录小人的失败位置
        if "_" not in board:#如果单词全部输对就就获胜，并且打印单词
            print("胜利!")
            print(" ".join(board))
            win = True
            break
    if not win:
        print("\n".join(HP[0:wrong+1]))#如果血量不足则失败
        print("您失败了,正确答案是{}".format(word))
sblsy = random.choice(dictionary)#随机抽取单词
hangman(sblsy)
```
